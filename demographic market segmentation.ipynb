```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
data=pd.read_excel("EV Demographic.xlsx")
```

```python
data.head()
```

```python
data.columns
```

```python
data.describe()
```

```python
data.drop("S.No",axis=1,inplace=True)
```

```python
data.plot.bar()
```

```python
data.head()
```

```python
l=list(data["Age "])
```

```python
l3=[]
for i in l:
  l2=i.split("-")
  l2=[int(a) for a in l2]
  m=int(sum(l2)/2)
  l3.append(m)
```

```python
data["Age "]=l3
```

```python
data.head()
```

```python
plt.figure(figsize=(10,10))
plt.subplot(2,2,1)
plt.title("Age")
plt.bar(data.iloc[:,0],data.iloc[:,1])
plt.xticks(rotation="vertical")
plt.subplot(2,2,2)
plt.title("Total Vehicles")
plt.bar(data.iloc[:,0],data.iloc[:,5])
plt.xticks(rotation="vertical")
plt.show()
```

```python
  
X_axis = np.arange(len(l3))
plt.figure(figsize=(15,10))  
plt.bar(X_axis - 0.2, data.iloc[:,2], 0.4, label = 'Female')
plt.bar(X_axis + 0.2, data.iloc[:,3], 0.4, label = 'Male')
  
plt.xticks(X_axis, data.iloc[:,0],rotation=60)
plt.xlabel("Groups")
plt.ylabel("Number of Person")
plt.title("Number of Person in each group")
plt.legend()
plt.show()
```
